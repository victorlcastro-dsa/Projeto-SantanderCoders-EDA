##PaÃ­s com Maior EmissÃ£o de COâ‚‚

```python
# Selecionar as variÃ¡veis de interesse
columns_country_CO2 = ['Country', 'CO2_Emissions_MT']

# Verificar se hÃ¡ dados faltantes
print(df[columns_country_CO2].isnull().sum())

# Agrupar por paÃ­s e calcular o total de emissÃµes de COâ‚‚
co2_by_country = df.groupby('Country')['CO2_Emissions_MT'].sum().reset_index()

# Ordenar os paÃ­ses pelas maiores emissÃµes de COâ‚‚
co2_by_country = co2_by_country.sort_values(by='CO2_Emissions_MT', ascending=False)

# Exibir o paÃ­s com o maior Ã­ndice de emissÃµes de COâ‚‚
top_co2_country = co2_by_country.iloc[0]

print(co2_by_country)
print(f"PaÃ­s com maior emissÃµes de COâ‚‚: {top_co2_country['Country']} com {top_co2_country['CO2_Emissions_MT']} MT")
```

##RelaÃ§Ã£o entre EmissÃµes de COâ‚‚ e Eventos ClimÃ¡ticos Extremos

```python
#Verificar a RelaÃ§Ã£o entre EmissÃµes de COâ‚‚ e Eventos ClimÃ¡ticos Extremos

# Selecionar as variÃ¡veis de interesse para eventos climÃ¡ticos
columns_country_CO2_events = ['Country', 'CO2_Emissions_MT', 'Extreme_Weather_Events']

# Agrupar por paÃ­s e calcular a soma das emissÃµes e eventos extremos
co2_and_events_by_country = df.groupby('Country').agg({
    'CO2_Emissions_MT': 'sum',
    'Extreme_Weather_Events': 'sum'
}).reset_index()

# Verificar a correlaÃ§Ã£o entre emissÃµes de COâ‚‚ e eventos extremos

numeric_columns = co2_and_events_by_country.select_dtypes(include=['number'])
correlation_CO2_events = co2_and_events_by_country[['CO2_Emissions_MT', 'Extreme_Weather_Events']].corr()

print(f"CorrelaÃ§Ã£o entre emissÃµes de COâ‚‚ e eventos extremos: {correlation_CO2_events}")


# Plot the correlation matrix
plt.figure(figsize=(6, 4))
sns.heatmap(correlation_CO2_events, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('correlaÃ§Ã£o entre as variÃ¡veis')
plt.show()
```
