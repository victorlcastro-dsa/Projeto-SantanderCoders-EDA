```python
# Filtrar dados para anos com e sem eventos climáticos extremos
extreme_events = df[df['Extreme_Weather_Events'] == 1]  # Eventos climáticos extremos
no_extreme_events = df[df['Extreme_Weather_Events'] == 0]  # Sem eventos climáticos extremos
```

```python
# Calcular médias para anos com eventos climáticos extremos
extreme_events_impact = extreme_events.agg({
    'Crop_Yield_MT_per_HA': 'mean',
    'Economic_Impact_Million_USD': 'mean'
})

# Calcular médias para anos sem eventos climáticos extremos
no_extreme_events_impact = no_extreme_events.agg({
    'Crop_Yield_MT_per_HA': 'mean',
    'Economic_Impact_Million_USD': 'mean'
})

# Exibir os resultados
print("Com Eventos Climáticos Extremos")
print(extreme_events_impact)

print("\nSem Eventos Climáticos Extremos")
print(no_extreme_events_impact)
```

##País com Maior Emissão de CO₂

```python
# Selecionar as variáveis de interesse
columns_country_CO2 = ['Country', 'CO2_Emissions_MT']

# Verificar se há dados faltantes
print(df[columns_country_CO2].isnull().sum())

# Agrupar por país e calcular o total de emissões de CO₂
co2_by_country = df.groupby('Country')['CO2_Emissions_MT'].sum().reset_index()

# Ordenar os países pelas maiores emissões de CO₂
co2_by_country = co2_by_country.sort_values(by='CO2_Emissions_MT', ascending=False)

# Exibir o país com o maior índice de emissões de CO₂
top_co2_country = co2_by_country.iloc[0]

print(co2_by_country)
print(f"País com maior emissões de CO₂: {top_co2_country['Country']} com {top_co2_country['CO2_Emissions_MT']} MT")
```

##Relação entre Emissões de CO₂ e Eventos Climáticos Extremos

```python
#Verificar a Relação entre Emissões de CO₂ e Eventos Climáticos Extremos

# Selecionar as variáveis de interesse para eventos climáticos
columns_country_CO2_events = ['Country', 'CO2_Emissions_MT', 'Extreme_Weather_Events']

# Agrupar por país e calcular a soma das emissões e eventos extremos
co2_and_events_by_country = df.groupby('Country').agg({
    'CO2_Emissions_MT': 'sum',
    'Extreme_Weather_Events': 'sum'
}).reset_index()

# Verificar a correlação entre emissões de CO₂ e eventos extremos

numeric_columns = co2_and_events_by_country.select_dtypes(include=['number'])
correlation_CO2_events = co2_and_events_by_country[['CO2_Emissions_MT', 'Extreme_Weather_Events']].corr()

print(f"Correlação entre emissões de CO₂ e eventos extremos: {correlation_CO2_events}")


# Plot the correlation matrix
plt.figure(figsize=(6, 4))
sns.heatmap(correlation_CO2_events, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('correlação entre as variáveis')
plt.show()
```

#####Conclusão

Com base na análise exploratória realizada, ficou evidente que há uma correlação positiva significativa (0.77) entre as emissões de CO₂ e a ocorrência de eventos climáticos extremos. A Austrália, com a maior emissão de CO₂ registrada no conjunto de dados (15889.61 MT), também sofreu 5192 eventos climáticos extremos, resultando em um impacto econômico expressivo de 682,067.12 milhões de dólares.

As estratégias de adaptação adotadas pela Austrália, como o uso de cultivos resistentes à seca, agricultura orgânica, gestão da água, e rotação de culturas, estão sendo implementadas para mitigar os danos causados pelas mudanças climáticas.

Conclui-se que o aumento das emissões de CO₂ contribui significativamente para o crescimento dos eventos climáticos extremos, reforçando a urgência de iniciativas que reduzam as emissões e promovam a adaptação, especialmente no setor agrícola.
